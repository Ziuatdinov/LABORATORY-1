```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


# Функция выполняет сложение изображений
def addImages(img1, img2):
    img2 = cv2.resize(img2, dsize=(img1.shape[1], img1.shape[0]))
    imgAdd = cv2.add(img1, img2)
    return imgAdd

# Функция выполняет вычитание изображений
def subtractImages(img1, img2):
    img2 = cv2.resize(img2, dsize=(img1.shape[1], img1.shape[0]))

    imgSubtract = cv2.subtract(img1, img2)
    return imgSubtract

# Загрузка изображений
img1 = cv2.imread('Images/ford.jpg')
img2 = cv2.imread('Images/img1.jpg')

img_add = addImages(img1,img2)  
img_subtract = subtractImages(img1,img2)      

flagAdd = True  # флаг сложения
flagSubtract = True # флаг вычитания

while True:
    img_show = img1

    if not flagAdd:
        img_show = img_add

    if not flagSubtract:
        img_show = img_subtract   
        
    cv2.imshow('Result', cv2.resize(img_show, dsize=(1000, 700)))

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    if key == ord('a'):
        flagSubtract = True
        flagAdd = not flagAdd
    if key == ord('s'):
        flagAdd = True
        flagSubtract = not flagSubtract

cv2.destroyAllWindows()
```

```python
import cv2

# Функция меняет кадр видео
def updateVideoAdd(img, img_add):
    img_add = cv2.resize(img_add, dsize=(img.shape[1], img.shape[0]))
    img = cv2.add(img, img_add)
    return img

def updateVideoSubtract(img, img_subtract):
    img_subtract = cv2.resize(img_subtract, dsize=(img.shape[1], img.shape[0]))
    img = cv2.subtract(img, img_subtract)
    return img

img_a = cv2.imread('Images/ford.jpg')  # фото мустанга
cap = cv2.VideoCapture(0)  # записываем вебку

flagVideoAdd = True  # флаг сложения
flagVideoSubtract = True # флаг вычитания

# вывод Opencv
while True:
    ret, frame = cap.read()

    if not flagVideoAdd:
        frame = updateVideoAdd(frame, img_a)

    if not flagVideoSubtract:
        frame = updateVideoSubtract(frame, img_a)   
        
    cv2.imshow('Camera', frame)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    if key == ord('a'):
        flagVideoSubtract = True
        flagVideoAdd = not flagVideoAdd
    if key == ord('s'):
        flagVideoAdd = True
        flagVideoSubtract = not flagVideoSubtract

cap.release()
cv2.destroyAllWindows()
```
